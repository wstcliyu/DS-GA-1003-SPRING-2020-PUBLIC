### JupyterLab

The platform provides a computing environment. Students can use Jupyter notebooks to combine code, text and charts.

 - **Usage**
    * Access JupyterLab in a Chrome browser with your NYU credentials. You must have enabled Multi-Factor Authentication. 
    * Under `Control Panel` you can start or stop your server. Note that `Logout` will not stop your server.
    * Sessions are terminated following a half hour of idleness or three hours of use. Please log in before lecture/section to avoid delays in your session.    
    * If you experience any issues, then please post to Piazza indicating your NetId. Your NetId will allow us to identify your server. 
 - **Organization**
    * Your folders should look like 
        - `shared/` : You have read access to the folder. We will add data to the folder needed for assignments.
        - `class_materials/from_github/` : Materials from lecture and sections including slides and notebooks. Note that the folder updates periodically with changes or additions. Any modifications to files in the folder will be overwritten.
        - `your_materials/`: Copies of the materials from lecture and section including slides and notebooks. You are encouraged to modify these files. Any modifications in the folder will be persisted throughout the semester.
    * Note that `class_materials/from_github/` and `your_materials/` fetch from [GitHub](https://github.com/cp71/DS-GA-1003-SPRING-2020-PUBLIC)
 - **Environments**
    * Package are installed under the environment `dsga-1003`. If you use the `base` environment instead of the `dsga-1003` environment you might have missing packages. 
    * If you need a package for your project, then please post to Piazza. 
 

### Piazza 

Please post any questions or comments on Piazza. If you have personal questions, then please use a private post instead of email. We are glad to help you resolve issues with the assignments, exams or logistics. However we need to enable you to solve problems. So before posting to Piazza:
 - Searching Piazza for existing posts. Use the folder/filter system
 - Read the documentation 
 - Searching the Web particularly [forums](https://stackoverflow.com/)
 - Experiment with different possibilities through trial and error informed by your experiences in class

We want to encourage students to help each other. Of course, we will still comment on logistics, typos, tricky material, etc. However we want should not give away the answers. We want to allow for discussion between students to foster collaboration. In order to make Piazza a better resource for everyone, we've outlined some guidelines for you to follow when posting your questions. Remember that we will increase your letter grade based on participation in class, section, office hour and Piazza. Questions and comments that follow these guidelines contribute to your participation.

1. Designated Posts
     * Ask assignment questions only in the designated posts. We've created individual posts for each problem. Please ask questions, discuss problems, or help out in those posts only. Before asking a question, read through (or search) the whole thread to see if your question has been answered.
2. No answers in Posts
    * Please don't give away the answer on Piazza. You can explain things in a way that still lets other students figure out the essence of the problem on their own, but don't spoil the problem. If you are not sure about the appropriateness of a post, then post privately to instructors and then we will inform you.
3. Try to make Posts Public
    * While not divulging answers, try to make your questions public because others might have the same question. 
4. Posts are not a substittute for Office Hours
    * If you think your questions may require lengthy answers, then please come to instructor, teaching assistant or grader office hours.
5. Posts are for grading.
    * Please do not post questions of the form: "Is this the correct solution to HW X problem Y?" or "Is this the right level of detail for hw X problem Y?" Please do not use Piazza as a channel for grading. Feel free to ask questions of clarification, or ask questions about the course content to achieve a deeper understanding, but at a certain point, you must apply your knowledge, give it your best shot, and submit your answers with confidence.
7. Posts should be self-contained
    * Please explain the steps in your reasoning:  "Starting out, we have: .... Then, I do ..., and I get ... Now, I get confused, and this makes no sense." Posting screen shots will help us to understand the bugs

##### Instructions for Formatting

We suggest three approaches to formatting your homework

 - _LaTex_
     * It would be helpful to use [Overleaf](https://www.overleaf.com/) for homework and project. The platform allow users to share projects amongst collaborators. 
     * You may find the [minted](https://github.com/gpoore/minted) package helpful for formatting code and the [algorithm2e](https://ctan.org/pkg/algorithm2e?lang=en) package helpful for formatting pseudo-code
 - _LyX_
    * If you would prefer a WYSIWYG version of Latex, then you could try LyX. 
    * If you are using LyX, then the [listings](https://en.wikibooks.org/wiki/LaTeX/Source_Code_Listings) package tends to work better for code.
 - _MathJax in Jupyter Notebook_
    * Jupyter Notebooks combine code, text and charts. You are able to embed Latex in markdown cells with either single or double dollar signs. See the [documentation](http://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Working%20With%20Markdown%20Cells.html) for more information and commands.
    * To convert to Portable Document Format please
        1. Download as HTML (`File->Download As->HTML(.html)`). 
        1. Open the HTML in the browser. Print to .pdf
    * The conversion requires [nbconvert](https://nbconvert.readthedocs.io/en/latest/install.html) along with dependencies (pandoc and TeX). Note that nbconvert is installed in the `dsga-1003` environment on JupyterLab.
    * Guidelines
        1. The generated HTML will be from the last save point so make sure you save all your changes before running the following cell.
        1. You should break long lines of code into multiple lines. Otherwise your code will extend out of view from the cell. Consider using `\` followed by a new line. 
        1. For each textual response, please include relevant code that informed your response. 
        1. For each plotting question, please include the code used to generate the plot. If your plot does not appear in the HTML / pdf output, then use `Image('name_of_file')` to embed it.
        1. You should not display large output cells such as all rows of a table. Instead convert the input cell from Code to Markdown back to Code to remove the output cell.

**Gradescope points will be awarded if and only if you followed all the submission and formatting instructions**
